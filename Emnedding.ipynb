### 1. What are Embedding ?

Embedding are numerical representations of text,image,audio , or any data. For text (usally our focus), an embedding is :

- A vector (a long list of numbers)

- Generated by a model

- Dedigned so that similar meanings -> similar numbers.

### 2.Why Do We Use Embedding ?

Because computer don't understand text directly-they understand numbers

Embedding allow systems to:

  - Search semantically ("meaning-based search")

  - Build recommendation systems

  - Measuer similarity between texts

  - Cluster documents

  - RAG (Retrival-Augmented Generation),including ChatGorq workflows

### 3. What Is a similarity Score ? 

Once you have embedding , you measure similarity using a mathematical formula

The most common one is :

**Cosine Similarity**

it measures the angel between two vectors.

    - 1.0 -> identical meaning

    - 0.0 -> unrelated

    - -1.0 -> opposite meaningm

```python
from sentence_transformers import SentenceTransformer
import numpy as np
```

```python
import warnings
warnings.filterwarnings('ignore')
model = SentenceTransformer("all-Minilm-L6-v2")
```

```python
sentences = [
    "I like apples.",
    "I enjoy eating apples.",
    "I dirve a motorcycle."
]
```

```python
embeddings = model.encode(sentences)
```

```python
def cosine_sim(a,b):
    return np.dot(a,b) / (np.linalg.norm(a) * np.linalg.norm(b))
```

```python
v1=embeddings[0]
v2=embeddings[1]
Nr=np.dot(v1,v2)
norm_v1=np.linalg.norm(v1)
norm_v2=np.linalg.norm(v2)
Dr=norm_v1*norm_v2
cosine_sim(v1,v2),Nr/Dr
```
